```python
import numpy as np
import random

# Make some non random Xk

list_Xk = []
nbXk = 10
for n in range(nbXk):
    random.seed(n)
    res = []
    for i in range(5):
        r = [random.random() for _ in range(5)]
        s = sum(r)
        res.append([ k/s for k in r ])
    res = np.array(res) 
    list_Xk.append(res)
    
print(list_Xk)
```

# Generate data

```python
from generatedata import generatelist

# Test with a given dictionary of sample
n = 1000
dk = generatelist(n, Xk='Condition_1', nb_replicates=10, mk=list_Xk[0]) 
```

```python
import pandas as pd
df = generatelist(n, Xk='Condition_1') 
seed = 2
for mat in list_Xk:
    
    name = 'Condition_' + str(seed)
    dk = generatelist(n, Xk=name, seed=seed, nb_replicates=10, mk=mat)
    df = pd.concat([df, dk[dk.columns[1:]]], axis=1)
    seed +=1
    
# Export data generated
import os
path = os.getcwd() + '\\data\\'
df.to_csv(path+'testdata1.csv')
```

```python
c = (df['Condition_1'] >= 80)
topExp = df[c].sort_values('Condition_1', ascending=False)

cprime = (df['Condition_1_prime1'] >= 80)
topExppr = df[cprime].sort_values('Condition_1_prime1', ascending=False)


l1 = list(topExp.index)
l1prime = list(topExppr.index)
print('_'*70)
print('score rbo=', rbo.rbo(l1, l1prime, p=0.1))
print('score rbo=', rbo.rbo_modified(rbo.wg_geom, params_wg_func={'geometric':None, 'p':0.1}, lstS=l1, lstT=l1prime))


print('score binomial=', rbo.rbo_modified(rbo.wg_binomial, params_wg_func={'binomial':None,
                                                                           'n': 20, 
                                                                           'p':0.5}, lstS=l1, lstT=l1prime))

print('_'*70)
print('score poisson=', rbo.rbo_modified(rbo.wg_poisson, params_wg_func={'poisson':None, 'lambda':10}, lstS=l1, lstT=l1prime))

print('_'*70)
K = len(set(l1).intersection(set(l1prime)))
N = max(len(set(l1)), len(set(l1prime)))

print('score hypergrom=', rbo.rbo_modified(rbo.wg_hypergeom, 
                                              params_wg_func={'hypergeometric':None, 
                                                              'N': N, 
                                                              'K': K, 
                                                              'n': 20},
                                              lstS=l1, lstT=l1prime))
```

```python
import string
import numpy as np
TESTS = [
    # Sanity checks
    (string.ascii_lowercase, string.ascii_lowercase, 1.0),
    (string.ascii_lowercase, string.ascii_lowercase[:7], 1.0),
    ("abcde", "fghij", 0.0),

    # RBO Paper Figure 5
    ("abcdefg", "zcavwxy", 0.312),

    # Source:  https://ragrawal.wordpress.com/2013/01/18/comparing-ranked-list/
    ("abcde", "bacde", 0.8),
    ("abcde", "abced", 0.95),

    # One-Element lists
    ("a", "a", 1.0),
    ("a", "b", 0),

    # Empty lists
    ("", "", 1),
    ("a", "", 0),
    ("", "a", 0),
]


def test_rbo(list_1: list, list_2: list, expected: float):
    """
    Args:
        list_1: List 1.
        list_2: List 2.
        expected: Expected RBO.
    Returns:
        None
    """
    p = 0.5  # pylint: disable=invalid-name
    list_1, list_2 = list(list_1), list(list_2)
    print("List 1 is: {}".format(list_1))
    print("List 2 is: {}".format(list_2))

    rbo_res = rbo.rbo(list_1, list_2)
    print(f"The implemented Average Overlap is: {rbo_res}")
    print(f"The correct answer is:              {expected}")
    assert np.round(rbo_res, decimals=3) == expected, "Not correct"
    
for test in TESTS:
    print(test_rbo(test[0], test[1], test[2]))
     
```
